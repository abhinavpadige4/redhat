```python
n = int(input("Enter value: "))
temp = n
fac_sum = 0
while temp > 0:
    digit = temp % 10
    temp //= 10
    fact = 1
    for i in range(1, digit + 1):
        fact *= i
    fac_sum += fact
if fac_sum == n:
    print(n)

```

```python
s = "sonoersd"
n = len(s)
max_len = 1

for i in range(n):
    l = i
    r = i
    while l >= 0 and r < n and s[l] == s[r]:
        max_len = max(max_len, r - l + 1)
        l -= 1
        r += 1
    l = i
    r = i + 1
    while l >= 0 and r < n and s[l] == s[r]:
        max_len = max(max_len, r - l + 1)
        l -= 1
        r += 1

print("Length of longest palindrome:", max_len)
```
