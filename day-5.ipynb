```python
s1=input("enter a word -1")
s2=input("enter a word -2")
print("palindrome" if s1==s2[::-1] else "not a palindrome")
```

```python
s1=input("enter a word -1")
s2=input("enter a word -2")
print("anagram" if sorted(s1)==sorted(s2) else "not aN anagram")
```

```python
word = input()
s = []  
l = []   

for ch in word:
    if ch in s:
        idx = s.index(ch)
        l[idx] += 1
    else:
        s.append(ch)
        l.append(1)

print(s, l)
```

```python
num=list(map(int,input()))
c=0
for i in num:
    x=sum(num)-i
    if i>x:
        c+=1
print(c)
```

```python
num=list(map(int,input()))
x=sum(num)//len(num)
c=0
for i in num:
    if i>x:
        c+=1
print(c)
```

```python
n = 4

for i in range(1, n + 1):
    for j in range(i, 0, -1):
        print(j, end="")
    for j in range(2, i + 1):
        print(j, end="")
    print()
```

```python
num=["a","a","b","a","c","c","c"]
i=0
while(i<len(num)-1):
    if(num[i]==num[i+1]):
        num.pop(i+1)
    else:
        i+=1
print(num)     
```

```python
s="hiiiiii i am abhinav"
m=0
words=s.split()
for i in words:
    x=len(i)
    if x>m:
        m=x
print(m)        
```

```python
s = "hiiiiii i am abhinav"
words = s.split()
m = 0

for i in words:
    if i == i[::-1]:   
        y = len(i)
        if y > m:
            m = y

print(m)
```
