```python
text = "a,b,c,d,a,b a"

for c in [",", ".", "!", "@", "'"]:
    text = text.replace(c, " ")

words = text.split()

f = {}
for w in words:
    if w in f:
        continue
    else:
        f[w] = 1

print(f)
```

```python

l = ["eat", "ate", "bat", "nat", "ant"]
f = {}
for i in l:
    x = "".join(sorted(i))
    if x not in f:
        f[x] = []
    f[x].append(i)

print(list(f.items()))
```
